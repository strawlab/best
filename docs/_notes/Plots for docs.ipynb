```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

# normality.png

```python
fig, ax = plt.subplots(figsize=(6, 4))
xs = np.linspace(-5, 5, 500)
for nu in [100, 10, 2.5]:
    ps = st.t(df=nu).pdf(xs)
    plt.plot(xs, ps, label=r'$\nu = %g$' % nu, alpha=0.7, lw=1.5)
plt.legend()
plt.xlabel('Observation')
plt.ylabel('Probability')
plt.ylim(0)
plt.xlim(-5, 5)
ax.text(-4.9, 0.03, 'Left tail', fontsize=12)
ax.text(4.9, 0.03, 'Right tail', fontsize=12, horizontalalignment='right')
fig.tight_layout()
plt.savefig('normality.png', dpi=180)
plt.show()
```
